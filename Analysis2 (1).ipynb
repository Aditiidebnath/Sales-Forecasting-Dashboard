```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Show all columns
pd.set_option("display.max_columns", None)

# Load the dataset
sales_file = os.path.join(sales_path, "train.csv")
sales_df = pd.read_csv(sales_file)

# Display basic information
print("Dataset Shape:", sales_df.shape)
display(sales_df.head())

# Check dataset information
print("\nDataset Information")
sales_df.info()

# Convert date columns (DD/MM/YYYY format)
sales_df["Order Date"] = pd.to_datetime(
    sales_df["Order Date"],
    format="%d/%m/%Y"
)

sales_df["Ship Date"] = pd.to_datetime(
    sales_df["Ship Date"],
    format="%d/%m/%Y"
)

# Create time features
sales_df["Year"] = sales_df["Order Date"].dt.year
sales_df["Month"] = sales_df["Order Date"].dt.month
sales_df["Month Name"] = sales_df["Order Date"].dt.month_name()
sales_df["Week Number"] = sales_df["Order Date"].dt.isocalendar().week
sales_df["Day of Week"] = sales_df["Order Date"].dt.day_name()
sales_df["Quarter"] = sales_df["Order Date"].dt.quarter

# Create season column
def get_season(month):
    if month in [12, 1, 2]:
        return "Winter"
    elif month in [3, 4, 5]:
        return "Spring"
    elif month in [6, 7, 8]:
        return "Summer"
    else:
        return "Autumn"

sales_df["Season"] = sales_df["Month"].apply(get_season)

# Check missing values
print("\nMissing Values")
print(sales_df.isnull().sum())

# Check missing value percentage
missing_percentage = (sales_df.isnull().sum() / len(sales_df)) * 100

print("\nMissing Value Percentage")
print(missing_percentage.round(2))

# Check duplicate rows
duplicates = sales_df.duplicated().sum()

print("\nDuplicate Rows:", duplicates)

duplicate_percentage = (duplicates / len(sales_df)) * 100

print("Duplicate Percentage:", round(duplicate_percentage, 2), "%")

# Remove duplicate rows
sales_df = sales_df.drop_duplicates()

# Check data types
print("\nData Types")
print(sales_df.dtypes)

# Create daily sales data
daily_sales = sales_df.groupby("Order Date")["Sales"].sum().reset_index()

print("\nDaily Sales")
display(daily_sales.head())

# Create weekly sales data
weekly_sales = daily_sales.resample(
    "W",
    on="Order Date"
)["Sales"].sum().reset_index()

print("\nWeekly Sales")
display(weekly_sales.head())

# Create monthly sales data
monthly_sales = daily_sales.resample(
    "ME",
    on="Order Date"
)["Sales"].sum().reset_index()

print("\nMonthly Sales")
display(monthly_sales.head())
# Analyze sales by product category

category_sales = sales_df.groupby("Category")["Sales"].sum().sort_values(ascending=False)

print("Total Sales by Category")
display(category_sales)

plt.figure(figsize=(7,5))
category_sales.plot(kind="bar", color="skyblue")
plt.title("Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.xticks(rotation=0)
plt.show()

print("Highest Revenue Category:", category_sales.idxmax())


# Analyze yearly sales by region

region_sales = sales_df.groupby(["Year", "Region"])["Sales"].sum().reset_index()

display(region_sales)

plt.figure(figsize=(10,6))

sns.lineplot(
    data=region_sales,
    x="Year",
    y="Sales",
    hue="Region",
    marker="o"
)

plt.title("Yearly Sales by Region")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.show()

# Create year-region table
region_pivot = region_sales.pivot(
    index="Year",
    columns="Region",
    values="Sales"
)

print("Year-wise Sales")
display(region_pivot)

# Calculate yearly growth
growth = region_pivot.pct_change() * 100

print("Year-over-Year Growth (%)")
display(growth.round(2))

# Find most consistent region
consistency = region_pivot.std().sort_values()

print("Sales Consistency")
display(consistency)

print("Most Consistent Region:", consistency.idxmin())


# Calculate shipping time

sales_df["Shipping Days"] = (
    sales_df["Ship Date"] -
    sales_df["Order Date"]
).dt.days

print("Average Shipping Time:", round(sales_df["Shipping Days"].mean(),2), "Days")

shipping_region = sales_df.groupby("Region")["Shipping Days"].mean()

print("Average Shipping Time by Region")
display(shipping_region)

plt.figure(figsize=(7,5))
shipping_region.plot(kind="bar", color="orange")
plt.title("Average Shipping Time by Region")
plt.xlabel("Region")
plt.ylabel("Days")
plt.xticks(rotation=0)
plt.show()


# Analyze monthly sales pattern

monthly_pattern = sales_df.groupby(
    ["Year", "Month Name"]
)["Sales"].sum().reset_index()

month_order = [
    "January","February","March","April",
    "May","June","July","August",
    "September","October","November","December"
]

pivot = monthly_pattern.pivot(
    index="Month Name",
    columns="Year",
    values="Sales"
)

pivot = pivot.reindex(month_order)

# Line chart
pivot.plot(figsize=(12,6), marker="o")
plt.title("Monthly Sales Across Years")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.show()

# Heatmap
plt.figure(figsize=(10,6))

sns.heatmap(
    pivot,
    annot=True,
    cmap="YlGnBu",
    fmt=".0f"
)

plt.title("Monthly Sales Heatmap")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()

# Average monthly sales
average_month = sales_df.groupby("Month Name")["Sales"].mean()
average_month = average_month.reindex(month_order)

plt.figure(figsize=(10,5))
average_month.plot(marker="o")
plt.title("Average Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Average Sales")
plt.grid(True)
plt.show()

print("Highest Average Sales Month:", average_month.idxmax())


# Display summary

print("\n========== Task 1 Summary ==========")
print("Dataset Shape:", sales_df.shape)
print("Missing Values:", sales_df.isnull().sum().sum())
print("Duplicate Rows Removed:", duplicates)
print("Highest Revenue Category:", category_sales.idxmax())
print("Most Consistent Region:", consistency.idxmin())
print("Average Shipping Time:", round(sales_df["Shipping Days"].mean(),2), "Days")
print("Highest Sales Month:", average_month.idxmax())
print("====================================")
```

**Task 2 — Time Series Analysis & Decomposition**

```python
# Import required libraries
!pip -q install statsmodels

from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller

# Create time series
sales_ts = monthly_sales.copy()
sales_ts.set_index("Order Date", inplace=True)

# Plot monthly sales
plt.figure(figsize=(12,5))

plt.plot(
    sales_ts.index,
    sales_ts["Sales"],
    marker="o",
    color="blue",
    linewidth=2
)

plt.title("Monthly Sales Trend")
plt.xlabel("Order Date")
plt.ylabel("Sales")
plt.grid(True)

plt.show()

# Apply time series decomposition
decomposition = seasonal_decompose(
    sales_ts["Sales"],
    model="additive",
    period=12
)

# Plot decomposition
fig = decomposition.plot()

fig.set_size_inches(14,10)

plt.show()

# Check stationarity using ADF test
print("ADF Test Result")

adf_result = adfuller(sales_ts["Sales"])

print("ADF Statistic :", round(adf_result[0],4))
print("P-value :", round(adf_result[1],4))

print("\nCritical Values")

for key, value in adf_result[4].items():
    print(key, ":", round(value,4))

if adf_result[1] < 0.05:
    print("\nThe time series is stationary.")
else:
    print("\nThe time series is not stationary.")

# Apply differencing
sales_diff = sales_ts["Sales"].diff().dropna()

# Plot differenced data
plt.figure(figsize=(12,5))

plt.plot(
    sales_diff,
    marker="o",
    color="green"
)

plt.title("Monthly Sales After Differencing")
plt.xlabel("Order Date")
plt.ylabel("Differenced Sales")
plt.grid(True)

plt.show()

# Check stationarity again
print("ADF Test After Differencing")

adf_diff = adfuller(sales_diff)

print("ADF Statistic :", round(adf_diff[0],4))
print("P-value :", round(adf_diff[1],4))

print("\nCritical Values")

for key, value in adf_diff[4].items():
    print(key, ":", round(value,4))

if adf_diff[1] < 0.05:
    print("\nThe differenced series is stationary.")
else:
    print("\nThe differenced series is still not stationary.")
```

## Task 2 – Observations

**Observation 1:**
The monthly sales trend shows an overall increasing pattern over the four-year period, indicating that the business experienced steady sales growth.

**Observation 2:**
The seasonal decomposition shows a repeating seasonal pattern, which indicates that sales are influenced by recurring monthly trends.

**Observation 3:**
 Larger spikes in the residual plot represent months with unusual sales activity compared to the expected pattern.

**Observation 4:**
The Augmented Dickey-Fuller (ADF) test produced an ADF Statistic of **-4.4161** with a **p-value of 0.0003**. Since the p-value is less than **0.05**, the monthly sales data is **stationary**, meaning its statistical properties remain relatively constant over time. After applying first-order differencing, the ADF Statistic became **-8.7271** with a **p-value of 0.0000**, confirming that the differenced series is also stationary and suitable for time series forecasting models such as SARIMA.


**Task 3 — Sales Forecasting using 3 Different Models**

**MODEL1: SARIMA**

```python
# Install required library
!pip -q install statsmodels

# Import required libraries
import numpy as np
import matplotlib.pyplot as plt

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error

from statsmodels.tsa.statespace.sarimax import SARIMAX

# Create time series
sales_ts = monthly_sales.copy()

sales_ts.set_index("Order Date", inplace=True)

# Split data into train and test
train = sales_ts.iloc[:-3]
test = sales_ts.iloc[-3:]

# Build SARIMA model
sarima_model = SARIMAX(
    train["Sales"],
    order=(1,1,1),
    seasonal_order=(1,1,1,12),
    enforce_stationarity=False,
    enforce_invertibility=False
)

# Train model
sarima_result = sarima_model.fit()

print(sarima_result.summary())

# Predict test data
sarima_pred = sarima_result.forecast(steps=3)

# Calculate evaluation metrics
mae_sarima = mean_absolute_error(test["Sales"], sarima_pred)
rmse_sarima = np.sqrt(mean_squared_error(test["Sales"], sarima_pred))

mape_sarima = np.mean(
    np.abs((test["Sales"] - sarima_pred) / test["Sales"])
) * 100

print("\nMAE :", round(mae_sarima,2))
print("RMSE :", round(rmse_sarima,2))
print("MAPE :", round(mape_sarima,2), "%")

# Forecast next 3 months
future_forecast = sarima_result.get_forecast(steps=3)

forecast_values = future_forecast.predicted_mean

confidence_interval = future_forecast.conf_int()

print("\nForecast for Next 3 Months")

display(forecast_values)

# Plot actual vs forecast
plt.figure(figsize=(12,6))

plt.plot(
    train.index,
    train["Sales"],
    label="Training Data"
)

plt.plot(
    test.index,
    test["Sales"],
    label="Actual Sales",
    marker="o"
)

plt.plot(
    test.index,
    sarima_pred,
    label="Predicted Sales",
    marker="o"
)

plt.fill_between(
    confidence_interval.index,
    confidence_interval.iloc[:,0],
    confidence_interval.iloc[:,1],
    color="lightgray",
    alpha=0.5,
    label="Confidence Interval"
)

plt.title("SARIMA Forecast")
plt.xlabel("Order Date")
plt.ylabel("Sales")
plt.legend()

plt.grid(True)

plt.show()
```

### SARIMA Model

The SARIMA model was trained using **order = (1,1,1)** and **seasonal order = (1,1,1,12)**. Here, **d = 1** was used to make the data stationary, while **m = 12** represents yearly seasonality in the monthly sales data. These parameters were selected as a simple and suitable configuration for forecasting monthly sales.


**MODEL2: Facebook Prophet (Industry-standard Forecasting Tool)**


```python
# Install required library
!pip -q install prophet

# Import required libraries
from prophet import Prophet
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
import numpy as np

# Prepare data for Prophet
prophet_data = monthly_sales.copy()

prophet_data.rename(
    columns={
        "Order Date": "ds",
        "Sales": "y"
    },
    inplace=True
)

# Split data into train and test
train = prophet_data.iloc[:-3]
test = prophet_data.iloc[-3:]

# Create Prophet model
prophet_model = Prophet(
    yearly_seasonality=True,
    weekly_seasonality=False,
    daily_seasonality=False
)

# Train model
prophet_model.fit(train)

# Create future dates
future = prophet_model.make_future_dataframe(
    periods=3,
    freq="ME"
)

# Generate forecast
forecast = prophet_model.predict(future)

# Display forecast
display(forecast[["ds","yhat","yhat_lower","yhat_upper"]].tail(3))

# Predict test data
test_pred = forecast["yhat"].iloc[-3:].values

# Calculate evaluation metrics
mae_prophet = mean_absolute_error(test["y"], test_pred)

rmse_prophet = np.sqrt(
    mean_squared_error(test["y"], test_pred)
)

mape_prophet = np.mean(
    np.abs((test["y"] - test_pred) / test["y"])
) * 100

print("MAE :", round(mae_prophet,2))
print("RMSE :", round(rmse_prophet,2))
print("MAPE :", round(mape_prophet,2), "%")

# Plot forecast
fig1 = prophet_model.plot(forecast)

plt.title("Prophet Forecast")
plt.grid(True)

plt.show()

# Plot trend and seasonality
fig2 = prophet_model.plot_components(forecast)

plt.show()

# Display forecast for next 3 months
print("Forecast for Next 3 Months")

display(
    forecast[["ds","yhat","yhat_lower","yhat_upper"]].tail(3)
)
```

### Prophet Model

The Prophet model was trained using the monthly sales data after converting the columns into **ds** (date) and **y** (sales), which is the required format for Prophet. The model automatically detected the trend and yearly seasonal pattern in the data and generated a forecast for the next three months. The forecast performance was evaluated using **MAE, RMSE, and MAPE**.


**MODEL3:XGBoost**

```python
# Install required library
!pip -q install xgboost

# Import required libraries
import xgboost as xgb
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error

# Create a copy of monthly sales
xgb_data = monthly_sales.copy()

# Create time features
xgb_data["Month"] = xgb_data["Order Date"].dt.month
xgb_data["Quarter"] = xgb_data["Order Date"].dt.quarter

# Create season feature
def get_season(month):
    if month in [12,1,2]:
        return 1
    elif month in [3,4,5]:
        return 2
    elif month in [6,7,8]:
        return 3
    else:
        return 4

xgb_data["Season"] = xgb_data["Month"].apply(get_season)

# Create lag features
xgb_data["Lag_1"] = xgb_data["Sales"].shift(1)
xgb_data["Lag_2"] = xgb_data["Sales"].shift(2)
xgb_data["Lag_3"] = xgb_data["Sales"].shift(3)

# Create rolling mean feature
xgb_data["Rolling_Mean_3"] = (
    xgb_data["Sales"]
    .rolling(3)
    .mean()
)

# Remove missing rows
xgb_data.dropna(inplace=True)

# Select features
features = [
    "Lag_1",
    "Lag_2",
    "Lag_3",
    "Rolling_Mean_3",
    "Month",
    "Quarter",
    "Season"
]

X = xgb_data[features]
y = xgb_data["Sales"]

# Split data
X_train = X.iloc[:-3]
X_test = X.iloc[-3:]

y_train = y.iloc[:-3]
y_test = y.iloc[-3:]

# Create model
xgb_model = xgb.XGBRegressor(
    n_estimators=100,
    learning_rate=0.1,
    max_depth=3,
    random_state=42
)

# Train model
xgb_model.fit(X_train, y_train)

# Predict
xgb_pred = xgb_model.predict(X_test)

# Calculate evaluation metrics
mae_xgb = mean_absolute_error(y_test, xgb_pred)

rmse_xgb = np.sqrt(
    mean_squared_error(y_test, xgb_pred)
)

mape_xgb = np.mean(
    np.abs((y_test - xgb_pred) / y_test)
) * 100

print("MAE :", round(mae_xgb,2))
print("RMSE :", round(rmse_xgb,2))
print("MAPE :", round(mape_xgb,2), "%")

# Forecast next 3 months
forecast_table = pd.DataFrame({
    "Actual Sales": y_test.values,
    "Predicted Sales": xgb_pred
})

print("\nForecast Results")
display(forecast_table)

# Plot actual vs predicted
plt.figure(figsize=(10,5))

plt.plot(
    y_test.values,
    marker="o",
    linewidth=2,
    label="Actual"
)

plt.plot(
    xgb_pred,
    marker="o",
    linewidth=2,
    label="Predicted"
)

plt.title("XGBoost Prediction")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.grid(True)

plt.show()
```

### XGBoost Model

The XGBoost model was trained using lag features, a 3-month rolling mean, and calendar-based features such as month, quarter, and season. These features help the model learn historical sales patterns and predict future monthly sales. The model performance was evaluated using MAE, RMSE, and MAPE.


**MODEL COMPARISION**

```python
# Create comparison table for all models

comparison = pd.DataFrame({
    "Model": ["SARIMA", "Prophet", "XGBoost"],

    "MAE": [
        round(mae_sarima, 2),
        round(mae_prophet, 2),
        round(mae_xgb, 2)
    ],

    "RMSE": [
        round(rmse_sarima, 2),
        round(rmse_prophet, 2),
        round(rmse_xgb, 2)
    ],

    "MAPE (%)": [
        round(mape_sarima, 2),
        round(mape_prophet, 2),
        round(mape_xgb, 2)
    ]
})

# Show comparison table
display(comparison)

# Find best model based on RMSE
best_model = comparison.loc[comparison["RMSE"].idxmin(), "Model"]

print("\nBest Model based on RMSE:", best_model)
```

**Task 4 — Product Category & Region Level Forecasting**

```python
# install xgboost
!pip -q install xgboost

import xgboost as xgb
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error

# convert order date to datetime
sales_df["Order Date"] = pd.to_datetime(sales_df["Order Date"], dayfirst=True)

# create monthly data for each segment
def create_data(df, column, value):
    temp = df[df[column] == value].copy()
    temp = temp.groupby(pd.Grouper(key="Order Date", freq="M"))["Sales"].sum().reset_index()
    return temp

# create features for model
def feature_engineering(df):

    df["Month"] = df["Order Date"].dt.month
    df["Quarter"] = df["Order Date"].dt.quarter

    df["Lag_1"] = df["Sales"].shift(1)
    df["Lag_2"] = df["Sales"].shift(2)
    df["Lag_3"] = df["Sales"].shift(3)

    df["Rolling"] = df["Sales"].rolling(3).mean()

    df.dropna(inplace=True)

    return df

# train xgboost model
def run_model(data, name):

    df = feature_engineering(data.copy())

    features = ["Lag_1","Lag_2","Lag_3","Rolling","Month","Quarter"]

    X = df[features]
    y = df["Sales"]

    # split data
    X_train = X.iloc[:-3]
    X_test = X.iloc[-3:]

    y_train = y.iloc[:-3]
    y_test = y.iloc[-3:]

    # model
    model = xgb.XGBRegressor(n_estimators=100, max_depth=3, learning_rate=0.1)
    model.fit(X_train, y_train)

    # prediction
    pred = model.predict(X_test)

    # print errors
    print("\n", name)
    print("MAE:", round(mean_absolute_error(y_test, pred),2))
    print("RMSE:", round(np.sqrt(mean_squared_error(y_test, pred)),2))

    # plot actual vs predicted
    plt.figure(figsize=(6,3))
    plt.plot(y_test.values, marker="o", label="Actual")
    plt.plot(pred, marker="o", label="Predicted")
    plt.title(name)
    plt.legend()
    plt.grid()
    plt.show()

    return np.mean(pred)

# ---------------------------
# CATEGORY FORECASTING (3 PLOTS)
# ---------------------------

furniture = create_data(sales_df, "Category", "Furniture")
technology = create_data(sales_df, "Category", "Technology")
office = create_data(sales_df, "Category", "Office Supplies")

furniture_fc = run_model(furniture, "Furniture Forecast")
technology_fc = run_model(technology, "Technology Forecast")
office_fc = run_model(office, "Office Supplies Forecast")

# ---------------------------
# REGION FORECASTING (2 PLOTS)
# ---------------------------

west = create_data(sales_df, "Region", "West")
east = create_data(sales_df, "Region", "East")

west_fc = run_model(west, "West Region Forecast")
east_fc = run_model(east, "East Region Forecast")

# ---------------------------
# COMPARISON CHART (1 PLOT)
# ---------------------------

labels = ["Furniture","Technology","Office Supplies","West","East"]
values = [furniture_fc, technology_fc, office_fc, west_fc, east_fc]

plt.figure(figsize=(8,4))
plt.bar(labels, values)
plt.title("Forecast Comparison")
plt.xticks(rotation=30)
plt.grid(axis="y")
plt.show()
```

**Task 5 — Anomaly Detection in Sales Data**

```python
# import required libraries
from sklearn.ensemble import IsolationForest
import numpy as np
import matplotlib.pyplot as plt

# convert order date to datetime
sales_df["Order Date"] = pd.to_datetime(sales_df["Order Date"], dayfirst=True)

# create weekly sales data
weekly_sales = sales_df.groupby(
    pd.Grouper(key="Order Date", freq="W")
)["Sales"].sum().reset_index()

# Isolation Forest method
model = IsolationForest(contamination=0.05, random_state=42)

weekly_sales["IF_Anomaly"] = model.fit_predict(weekly_sales[["Sales"]])

weekly_sales["IF_Anomaly"] = weekly_sales["IF_Anomaly"].apply(lambda x: 1 if x == -1 else 0)

# Z-score method
mean = weekly_sales["Sales"].mean()
std = weekly_sales["Sales"].std()

weekly_sales["Z_Score"] = (weekly_sales["Sales"] - mean) / std

weekly_sales["Z_Anomaly"] = weekly_sales["Z_Score"].apply(lambda x: 1 if abs(x) > 2 else 0)

# plot Isolation Forest anomalies
plt.figure(figsize=(12,5))

plt.plot(weekly_sales["Order Date"], weekly_sales["Sales"], label="Sales")

plt.scatter(
    weekly_sales[weekly_sales["IF_Anomaly"] == 1]["Order Date"],
    weekly_sales[weekly_sales["IF_Anomaly"] == 1]["Sales"],
    color="red",
    label="Anomaly"
)

plt.title("Isolation Forest Anomaly Detection")
plt.legend()
plt.grid()
plt.show()

# plot Z-score anomalies
plt.figure(figsize=(12,5))

plt.plot(weekly_sales["Order Date"], weekly_sales["Sales"], label="Sales")

plt.scatter(
    weekly_sales[weekly_sales["Z_Anomaly"] == 1]["Order Date"],
    weekly_sales[weekly_sales["Z_Anomaly"] == 1]["Sales"],
    color="orange",
    label="Anomaly"
)

plt.title("Z-Score Anomaly Detection")
plt.legend()
plt.grid()
plt.show()

# compare results
comparison = pd.DataFrame({
    "Method": ["Isolation Forest", "Z-Score"],
    "Total Anomalies": [
        weekly_sales["IF_Anomaly"].sum(),
        weekly_sales["Z_Anomaly"].sum()
    ]
})

display(comparison)
```

This indicates that the anomalies in the sales data are significant enough to be detected by both machine learning-based and statistical methods.

It also shows that the dataset contains consistent spike and drop patterns, likely caused by real-world events such as promotions, festivals, or sudden demand changes.

**Task 6 — Product Demand Segmentation using Clustering**


```python
# import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# convert order date
sales_df["Order Date"] = pd.to_datetime(sales_df["Order Date"], dayfirst=True)

# feature engineering

# total sales per sub-category per year
yearly = sales_df.groupby([sales_df["Order Date"].dt.year, "Sub-Category"]).agg({
    "Sales": "sum"
}).reset_index()

yearly.columns = ["Year", "Sub-Category", "Sales"]

# total sales volume
total_sales = sales_df.groupby("Sub-Category")["Sales"].sum()

# average order value
avg_order = sales_df.groupby("Sub-Category")["Sales"].mean()

# sales volatility (std dev)
volatility = sales_df.groupby("Sub-Category")["Sales"].std()

# sales growth rate (YoY)
growth = yearly.pivot(index="Sub-Category", columns="Year", values="Sales").fillna(0)

growth_rate = (growth.iloc[:, -1] - growth.iloc[:, 0]) / (growth.iloc[:, 0] + 1)

# combine features
cluster_data = pd.DataFrame({
    "Total_Sales": total_sales,
    "Avg_Order_Value": avg_order,
    "Volatility": volatility,
    "Growth_Rate": growth_rate
}).fillna(0)

# scaling

scaler = StandardScaler()
scaled_data = scaler.fit_transform(cluster_data)

# elbow method

wcss = []

for i in range(1, 10):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(scaled_data)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(6,4))
plt.plot(range(1,10), wcss, marker="o")
plt.title("Elbow Method")
plt.xlabel("Clusters")
plt.ylabel("WCSS")
plt.grid()
plt.show()

# KMeans clustering

kmeans = KMeans(n_clusters=3, random_state=42)
cluster_data["Cluster"] = kmeans.fit_predict(scaled_data)

# PCA for visualization

pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_data)

cluster_data["PCA1"] = pca_result[:,0]
cluster_data["PCA2"] = pca_result[:,1]

plt.figure(figsize=(7,5))

plt.scatter(
    cluster_data["PCA1"],
    cluster_data["PCA2"],
    c=cluster_data["Cluster"],
    cmap="viridis"
)

for i in cluster_data.index:
    plt.text(cluster_data["PCA1"][i], cluster_data["PCA2"][i], i, fontsize=8)

plt.title("Product Demand Segmentation")
plt.grid()
plt.show()

# show final table
display(cluster_data)
```

###TASK 6 Question:
**What stocking strategy would you recommend for each cluster?**

**Answer:**
The clustering results divide the products into three main demand groups based on total sales, average order value, volatility, and growth rate.

**Cluster 0** (Premium Demand):
This cluster mainly contains Copiers, which show very high sales and strong performance. These products should be stocked carefully with high priority due to their high revenue contribution.

**Cluster 1** ( Core Products):
This cluster includes Accessories, Binders, Chairs, Phones, Storage, Tables, and similar items. These are the main revenue-generating products and should be maintained with consistent and high inventory levels.

**Cluster 2** (Low Priority Products):
This cluster includes Art, Envelopes, Fasteners, Labels, Paper, and similar low-demand items. These products should be stocked in limited quantities to avoid overstocking and storage cost issues.

Overall, the clustering helps optimize inventory planning by separating high-value, core, and low-demand products for better business decision making.
